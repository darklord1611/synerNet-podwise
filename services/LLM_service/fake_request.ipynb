```python
from supabase import create_client, Client
from config import SUPABASE_URL, SUPABASE_KEY
import requests

supabase: Client = create_client(SUPABASE_URL, SUPABASE_KEY)
```

```python
import json

def load_json(path):
    with open(path) as f:
        return json.load(f)

episode_id = 54
suffix = "2h"

summary = load_json(f'./output/temp_summary_{suffix}.json')
keywords = load_json(f'./output/temp_keywords_{suffix}.json')
highlights = load_json(f'./output/temp_highlights_{suffix}.json')
keypoints = load_json(f'./output/temp_keypoints_{suffix}.json')
```

```python
response = (supabase.table('episode_summaries').insert([
    {
        'id': episode_id,
        'summary': summary,
        'keywords': keywords,
        'highlights': highlights,
        'keypoints': keypoints,
    }
]).execute())
response
```
